# One Hot Encoding using loop

```python
columns1 = ['sex']
columns2 = ['smoker']
columns3 = ['region']
for col1 in columns1:
    one_hot1 = pd.get_dummies(df6[col1],drop_first=True,prefix='dummy')
   

for col2 in columns2:
    one_hot2 = pd.get_dummies(df6[col2],drop_first=True,prefix='dummy')
    

for col3 in columns3:
    one_hot3 = pd.get_dummies(df6[col3],drop_first=True,prefix='dummy')
    
    
    new_df6 = pd.concat([df6,one_hot1,one_hot2,one_hot3],axis=1)
```

# Ordinal Encoding

```python
gender = df7.sex.unique()
habit = df7.smoker.unique()
location = df7.region.unique()
```

```python
gender
```

```python
habit
```

```python
location
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordi1 = OrdinalEncoder(categories=[gender])
ordi2 = OrdinalEncoder(categories=[habit])
ordi3 = OrdinalEncoder(categories=[location])
```

```python
ordi1
```

```python
encoded_value1 = ordi1.fit_transform(df7[['sex']])
encoded_value2 = ordi2.fit_transform(df7[['smoker']])
encoded_value3 = ordi3.fit_transform(df7[['region']])
```

```python
encoded_value1
```

```python
encoded_value2
```

```python
encd1 = pd.DataFrame(encoded_value1 , columns =['sex_encd_value'])
encd2 = pd.DataFrame(encoded_value2 , columns =['smoker_encd_value'])
encd3 = pd.DataFrame(encoded_value3 , columns =['region_encd_value'])
```

```python
encd1
```

```python
df7['sex'] = encd1
df7['smoker'] = encd2
df7['region'] = encd3
```
